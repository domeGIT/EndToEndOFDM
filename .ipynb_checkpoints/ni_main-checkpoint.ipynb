```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
from scipy.interpolate import interp1d


import scipy.interpolate
```

# Podnosioci

Broj pilota \( P \) u OFDM simbolu opisuje koliko se podnosilaca koristi za prenos poznatih informacija (tj. pilota). Piloti će se na prijemnoj strani koristiti za procenu bežičnog kanala između predajnika i prijemnika. Takođe, definišemo i vrednost koju svaki pilot prenosi (koja je poznata prijemniku).

Sada ćemo definisati neke skupove indeksa koji opisuju koji nosioci prenose pilote, a koji nosioci sadrže korisne podatke.

```python
K = 64 # broj podnosioca
CP = K//4  # duzina ciklicnog prefiksa
P = 8 # broj pilot podnosioca po OFDM bloku
pilotValue = 3+3j # vrednost pilot signala 


allCarriers = np.arange(K)  # indeksi svih podnosioca ([0, 1, ..., K-1])
pilotCarriers = allCarriers[::K//P] # Piloti su svaki (K/P)-ti podnosilac.

# Radi lakse procene kanala, neka poslednji nosioci takodje budu piloti.
pilotCarriers = np.hstack([pilotCarriers, np.array([allCarriers[-1]])])
P = P+1

# Nosici podataka su svi preostali nosioci.
dataCarriers = np.delete(allCarriers, pilotCarriers)


plt.figure(figsize=(8,0.8))
plt.plot(pilotCarriers, np.zeros_like(pilotCarriers), 'bo', label='piloti')
plt.plot(dataCarriers, np.zeros_like(dataCarriers), 'ro', label='podaci')
plt.legend(fontsize=10, ncol=2)
plt.xlim((-1,K)); plt.ylim((-0.1, 0.3))
plt.xlabel('Podnosioci')
plt.yticks([])
plt.grid(True);
```

# 16 QAM konstelacija

Definišimo indeks modulacije $\mu$ i odgovarajuću tabelu mapiranja. Posmatramo prenos sa 16QAM modulacijom, što znači da imamo $\mu=4$ bita po simbolu. Takođe, mapiranje grupa od po 4 bita u simbole 16QAM konstelacije biće definisano u `mapping_table`.Koristimo Grejovo mapiranje  tj. dva susedna simbola u konstelaciji razlikuju se samo u jednom bitu, stop pomaže u obradi grešaka, s obzirom da najveći broj grešaka u detekciji čine "susedne greške".

```python
mu = 4 # bitova po simbolu (tj. 16QAM)
payloadBits_per_OFDM = len(dataCarriers)*mu  # broj korisnih  bitova po OFDM simbolu

mapping_table = {
    (0,0,0,0) : -3-3j,
    (0,0,0,1) : -3-1j,
    (0,0,1,0) : -3+3j,
    (0,0,1,1) : -3+1j,
    (0,1,0,0) : -1-3j,
    (0,1,0,1) : -1-1j,
    (0,1,1,0) : -1+3j,
    (0,1,1,1) : -1+1j,
    (1,0,0,0) :  3-3j,
    (1,0,0,1) :  3-1j,
    (1,0,1,0) :  3+3j,
    (1,0,1,1) :  3+1j,
    (1,1,0,0) :  1-3j,
    (1,1,0,1) :  1-1j,
    (1,1,1,0) :  1+3j,
    (1,1,1,1) :  1+1j
}
for bits in itertools.product([0, 1], repeat=4):
    Q = mapping_table[bits]
    plt.plot(Q.real, Q.imag, 'bo')
    plt.text(Q.real, Q.imag + 0.2, f"{bits[0]}{bits[1]}{bits[2]}{bits[3]}", 
             ha='center')

plt.grid(True)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.xlabel('I')
plt.ylabel('Q')
plt.title('16 QAM Konstelacija sa Gray Mapping')
plt.show()
```

```python
demapping_table = {v : k for k, v in mapping_table.items()} # tabela demapiranja
```

# Kanal
Sada ćemo definisati bežični kanal između predajnika i prijemnika. Ovde koristimo  višeputni kanal sa zadatim impulsivnim odzivom `channelResponse`. Takođe, prikazujemo odgovarajući frekvencijski odziv. Kao što vidimo, kanal je frekvencijski selektivan. 

```python
channelResponse = np.array([1, 0, 0.3+0.3j])  # impulsni odziv
H_ = np.fft.fft(channelResponse, K)
plt.plot(allCarriers, abs(H_))
plt.xlabel('Podnosioci') 
plt.ylabel('amplitudni spektar') 
plt.grid(True) 
plt.xlim(0, K-1)

SNRdb = 25  # odnos signal-sum u dB koji bi trebalo da se pojavi na prijemniku. 
```

# OFDM 
Sve počinje nasumičnim nizom bitova.  


```python
bits = np.random.randint(0, 2, payloadBits_per_OFDM)
print ("broj bitova: ", len(bits))
print ("prvih dvadeset: ", bits[:20])
```

#### Serijsko-paralelni konvertor
Bitovi se šalju u serijsko-paralelni konvertor, koji grupiše bitove za OFDM frejm u grupe od \($\mu$) bita (tj. po jedna grupa za svaki podnosioc):

```python
bits_SP = bits.reshape((len(dataCarriers), mu))
print ("prvih 5 grupa:")
print (bits_SP[:5,:])
```

### Mapiranje
Sada se grupe bitova šalju u maper, koji ih transformiše u kompleksne simbole konstalacije uz pomoć `mapping_table`.

```python
# Mapiraj svaki niz od 4 bita u odgovarajuci QAM simbol
QAM_symbols = []
for bit_sequence in bits_SP:
     symbol_key = tuple(bit_sequence)
     qam_symbol = mapping_table[symbol_key]
     QAM_symbols.append(qam_symbol)

QAM = np.array(QAM_symbols)
print("prvih 5 QAM simbola i bitova:")
print(bits_SP[:5, :])
print(QAM[:5])
```

### Dodela podataka i pilota odgovarajućim podnosiocima
Za svaki podnosioc smo već definisali da li nosi podatke ili pilote pomoću nizova `dataCarriers` i `pilotCarriers`. Sada, da bismo formirali kompletni OFDM simbol, moramo da rasporedimo podatke i pilote na odgovarajuće OFDM podnosiocе:

```python
# Kreiraj OFDM simbol direktno
OFDM_data = np.zeros(K, dtype=complex)
OFDM_data[pilotCarriers] = pilotValue    # Postavi pilote
OFDM_data[dataCarriers] = QAM            # Postavi korisne podatke
```

### IDFT
Sada se OFDM podnosioci mogu transformisati u vremenski domen primenom IDFT. 

```python
OFDM_time = np.fft.ifft(OFDM_data)
```

Ciklični prefiks se dodaje na početak svakog OFDM simbola kako bi se eliminisali efekti intersimbolske interferencije  i omogućila jednostavnija ekvalizacija. 

```python
# Dodaj CP sa jasnim prikazom strukture
cyclic_prefix = OFDM_time[-CP:]  # ZAŠTITNI DEO (kopija kraja)
ofdm_body = OFDM_time           # GLAVNI DEO

OFDM_withCP = np.hstack([cyclic_prefix, ofdm_body])

print("OFDM blok:")
print(f"┌{'─'*20}┬{'─'*30}┐")
print(f"│ Ciklicni prefiks ({CP}) │ OFDM simbol ({len(OFDM_time)}) │")
print(f"└{'─'*20}┴{'─'*30}┘")
print(f"Ukupna duzina OFDM bloka: {len(OFDM_withCP)}")
```

### Put od predajnika do prijemnika
Sada se signal šalje na antenu i prenosi kroz vazduh do prijemnika. Između dveju antena nalazi se bežični kanal (opisan ranije).Stoga je signal na prijemnoj anteni jednak konvoluciji predajnog signala sa impulsnim odzivom kanala. Dodatno, na signal dodajemo šum u skladu sa zadatom vrednošću SNR-a:

```python
# Simulacija bezicnog kanala
OFDM_TX = OFDM_withCP

# 1. Konvolucija sa impulsnim odzivom kanala
OFDM_after_channel = np.convolve(OFDM_TX, channelResponse)

# 2. Izracunaj snagu signala za SNR
received_power = np.mean(np.abs(OFDM_after_channel)**2)
noise_power = received_power * 10**(-SNRdb/10)


# 3. Dodaj AWGN sum
complex_noise = np.sqrt(noise_power/2) * (
    np.random.randn(len(OFDM_after_channel)) + 
    1j*np.random.randn(len(OFDM_after_channel))
)
OFDM_RX = OFDM_after_channel + complex_noise

# Vizuelizacija
plt.figure(figsize=(10, 3))
plt.plot(np.abs(OFDM_TX), 'b-', alpha=0.7, label='Predato')
plt.plot(np.abs(OFDM_RX), 'r-', alpha=0.7, label='Primljeno + sum')
plt.legend()
plt.title('Prenos OFDM signala bezicnim kanalom')
plt.xlabel('Vreme') 
plt.ylabel('Amplituda')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Sada, na prijemnoj strani, ciklički prefiks (CP) se uklanja iz signala i izdvaja se prozor od \( K \) odbiraka iz primljenog signala.Nakon toga, signal se vraća u frekvencijski domen kako bismo dobili primljenu vrednost na svakom podnosiocu.

```python
OFDM_RX_noCP = OFDM_RX[CP:(CP+K)]
OFDM_demod = np.fft.fft(OFDM_RX_noCP)
```

### Procena bezicnog kanala
Kao sledeći korak, potrebno je izvršiti procenu bežičnog kanala. U ilustrativne svrhe koristimo jednostavnu procenu kanala metodom „nulte sile“ (zero-forcing) praćenu jednostavnom interpolacijom. Princip procene kanala je sledeći:

Predajni signal sadrži pilotske vrednosti na određenim pilotskim nosiocima. Ove pilotske vrednosti i njihova pozicija u frekvencijskom domenu (tj. indeks pilotskog nosioca) poznati su prijemniku. Na osnovu primljenih informacija na pilotskim podnosiocima, prijemnik može da proceni uticaj bežičnog kanala na taj podnosioc (jer zna šta je poslato, a šta je primljeno). Na taj način, prijemnik dobija informaciju o karakteristikama kanala na pilotskim nosiocima. Međutim, on želi da zna šta se dogodilo na nosiocima sa korisnim podacima. Da bi to postigao, interpolira vrednosti kanala između pilotskih nosilaca kako bi dobio procenu kanala i na nosiocima sa podacima.

```python
def channelEstimate(OFDM_demod):
    # 1. Uzmi pilot vrednosti
    pilots_received = OFDM_demod[pilotCarriers]
    
    # 2. Izracunaj odziv kanala na pilotima
    H_pilots = pilots_received / pilotValue
    
    # 3. Interpoliraj magnitudu i fazu
    magnitude_est = interp1d(pilotCarriers, np.abs(H_pilots), kind='linear')(allCarriers)
    phase_est = interp1d(pilotCarriers, np.angle(H_pilots), kind='linear')(allCarriers)
    
    # 4. Rekonstuisi kompleksni odziv
    H_est = magnitude_est * np.exp(1j * phase_est)
    
    # Prikazi rezultate
    plt.plot(allCarriers, np.abs(H_), 'g-', label='Pravi kanal')
    plt.stem(pilotCarriers, np.abs(H_pilots), 'ro', label='Pilot estimacije')
    plt.plot(allCarriers, np.abs(H_est), 'b--', label='Estimiran kanal')
    plt.grid(True); plt.legend(); plt.ylim(0, 2)
    
    return H_est
Hest = channelEstimate(OFDM_demod) 
```

Sada kada je kanal procenjen na svim nosiocima, možemo da iskoristimo tu informaciju u koraku ekvalizacije kanala. Ovde se, za svaki podnosioc, uklanja uticaj kanala kako bismo povratili jasne (samo zašumljene) simbole konstelacije.

```python
equalized_Hest = OFDM_demod / Hest
```

Sledeći korak je odbacivanje pilotskih nosioca iz ekvalizovanog signala, jer oni ne sadrže korisne informacije, već su korišćeni isključivo u procesu procene kanala.Na taj nacin mozemo dobiti konstelacioni dijagram na prijemniku

```python
QAM_est = equalized_Hest[dataCarriers]
plt.plot(QAM_est.real, QAM_est.imag, 'bo');
plt.grid(True)
plt.xlabel('I')
plt.ylabel('Q')
plt.title("Konstelacija na prijemniku")
```

### Demapiranje
Sada kada smo povratili simbole konstelacije, potrebno je da pošaljemo ove kompleksne vrednosti u demaper kako bismo pretvorili tačke konstelacije u grupe bitova. To radimo tako sto poredimo dobijene tacke sa svim tackama konstelacije i onda biramo najblizu. 

```python
def Demapping(QAM_symbols):
    # Svi moguci konstelacioni simboli
    constellation_points = np.array(list(demapping_table.keys()))
    
    # Izracunaj udaljenosti svakog primljenog simbola do svake konstelacione tacke
    distances = np.abs(QAM_symbols.reshape(-1, 1) - constellation_points.reshape(1, -1))
    
    # Pronadji najblizu konstelacionu tacku za svaki simbol
    nearest_indices = distances.argmin(axis=1)
    hard_decision_symbols = constellation_points[nearest_indices]
    
    # Konvertuj simbole natrag u bitove
    estimated_bits = np.vstack([demapping_table[symbol] for symbol in hard_decision_symbols])
    
    return estimated_bits, hard_decision_symbols

PS_est, hardDecision = Demapping(QAM_est)
for qam, hard in zip(QAM_est, hardDecision):
    plt.plot([qam.real, hard.real], [qam.imag, hard.imag], 'b-o');
    plt.plot(hardDecision.real, hardDecision.imag, 'ro')
plt.grid(True) 
plt.xlabel('I')
plt.ylabel('Q')
plt.title('Hard demapiranje')
```

Na dijagramu iznad, plave tačke predstavljaju primljene QAM simbole, dok su crvene tačke povezane s njima najbliže moguće tačke iz konstelacije. Grupa bitova koja odgovara tim crvenim tačkama se vraća kao rezultat demapiranja.

Konačno, ove grupe bitova moraju da se pretvore u serijski niz bitova pomoću paralelno-serijske konverzije.Sada kada su svi bitovi dekodirani, izračunajmo stopu grešaka po bitovima (Bit Error Rate – BER):

```python
bits_est = PS_est.reshape((-1,))
print ("BER:", np.sum(abs(bits-bits_est))/len(bits))
```
